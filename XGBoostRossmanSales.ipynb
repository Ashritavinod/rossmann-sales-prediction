```python
def split_date(df):
  df['Date']=pd.to_datetime(df['Date'])
  df['Year']=df.Date.dt.year
  df['Month']=df.Date.dt.month
  df['Day']=df.Date.dt.day
  df['WeekOfYear']=df.Date.dt.isocalendar().week
```

Calculating since how many months the competion has been open

```python
def comp_month(df):
  df['CompetitionOpen']=12*(df.Year-df.CompetitionOpenSinceYear)+(df.Month-df.CompetitionOpenSinceMonth)
  df['CompetitionOpen']=df['CompetitionOpen'].map(lambda x: 0 if x<0 else x).fillna(0)
```

```python
comp_month(merged_df)
comp_month(merged_test_df)
```

```python
merged_df
```

Calculating the month since promo2 was open and calculating whether a new round of promotion has started in the current month.

```python
def promo_month(row):
  month2str={1:'Jan',2:'feb',3:'Mar',4:'Apr',5:'May',6:'Jun',7:'Jul',8:'Aug',9:'Sep',10:'Oct',11:'Nov',12:'Dec'}
  try:
    months=(row['PromoInterval'] or '').split(',')
    if row['Promo2Open'] and month2str[row['Month']] in months:
      return 1
    else:
      return 0
  except Exception:
    return 0
def promo_cols(df):
  df['Promo2Open']=12*(df.Year-df.Promo2SinceYear)+(df.WeekOfYear-df.Promo2SinceWeek)
  #since how long the promo has started
  df['Promo2Open']=df['Promo2Open'].map(lambda x: 0 if x<0 else x).fillna(0)*df['Promo2']
  #is promotion happening in the current month
  df['IsPromo2Month']=df.apply(promo_month,axis=1)*df['Promo2']
```

```python
promo_cols(merged_df)
promo_cols(merged_test_df)
```

Scale Numeric Values

```python
from sklearn.preprocessing import MinMaxScaler
scaler=MinMaxScaler()
X[numeric_cols]=scaler.fit_transform(X[numeric_cols])
X_test[numeric_cols]=scaler.fit_transform(X_test[numeric_cols])
```

Encode Categoricol columns

```python
from sklearn.preprocessing import OneHotEncoder
import pandas as pd

# Convert all columns in categoricol_cols to string type before fitting the encoder
X[categoricol_cols] = X[categoricol_cols].astype(str)
X_test[categoricol_cols] = X_test[categoricol_cols].astype(str)  # Ensure X_test types match X

# Now fit the encoder on the combined data
combined_data = pd.concat([X[categoricol_cols], X_test[categoricol_cols]])  # Combine for fitting
encoder = OneHotEncoder(handle_unknown='ignore', sparse_output=False).fit(combined_data) #Fit on combined data
encoded_cols = list(encoder.get_feature_names_out(categoricol_cols))

# Transform both X and X_test
X_encoded = pd.DataFrame(encoder.transform(X[categoricol_cols]), columns=encoded_cols, index=X.index)
X_test_encoded = pd.DataFrame(encoder.transform(X_test[categoricol_cols]), columns=encoded_cols, index=X_test.index)

# Concatenate the encoded columns with the original DataFrames
X = pd.concat([X, X_encoded], axis=1)
X_test = pd.concat([X_test, X_test_encoded], axis=1)
```

K-fold cross validation

```python
from sklearn.model_selection import KFold
```

```python
from sklearn.metrics import root_mean_squared_error as rmse
def train_and_evaluate(X_train,y_train,X_val,y_val,**params):
  model=XGBRegressor(random_state=42,n_jobs=-1,**params)
  model.fit(X_train,y_train)
  train_rmse=rmse(model.predict(X_train),y_train)
  val_rmse=rmse(model.predict(X_val),y_val)
  return model,train_rmse,val_rmse
```

```python
KFold=KFold(n_splits=5)
```

```python
models=[]
for train_idxs,val_idxs in KFold.split(X):
  X_train,y_train=X.iloc[train_idxs],y.iloc[train_idxs]
  X_val,y_val=X.iloc[val_idxs],y.iloc[val_idxs]
  model,train_rmse,val_rmse=train_and_evaluate(X_train,y_train,X_val,y_val,max_depth=4,n_estimators=20)
  models.append(model)
  print("Train RMSE:{},Validation RMSE:{}".format(train_rmse,val_rmse))
```

```python
import numpy as np
def predict_avg(models,inputs):
  return np.mean([model.predict(inputs) for model in models],axis=0)
```

```python
preds=predict_avg(models,X)
```
